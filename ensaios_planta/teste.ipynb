```python
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
import control as ct
import scienceplots

#Exibição dos gráficos em janela externa:
#%matplotlib qt

# Configurações de plot:
plt.style.use([
    'grid',
    'retro'
])
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['font.size'] = 12 
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
thxvt = np.loadtxt('new_data\\thxvt.txt')           #velocidade  ângulo
thxvt = [thxvt[:, 0], thxvt[:, 1]]
pmot = np.loadtxt('pontos-motor.txt')           #velocidade  ângulo
pmot = [pmot[:, 0], pmot[:, 1], pmot[:, 2]]
```

```python
vt = np.delete(pmot[0], 1)
w = np.delete(pmot[2], 1)

pol = np.polyfit(vt, w, 2)

vlin = np.linspace(vt[0], vt[-1], 100)

plt.figure()
plt.scatter(vt, w)
plt.plot(vlin, np.polyval(pol, vlin), linestyle='dashed', color='r')
plt.show()
```

```python
th = thxvt[0]
vt2 = thxvt[1]

w2 = np.polyval(pol, vt2)

plt.figure()
plt.scatter(w2, th)
plt.show()
```
